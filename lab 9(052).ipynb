```python
import csv

class Document:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}")


class Book(Document):
    def __init__(self, title, author, publisher=None, year=None):
        super().__init__(title, author)
        self.publisher = publisher
        self.year = year

    def display_info(self):
        info = f"Title: {self.title}, Author: {self.author}"
        if self.publisher and self.year:
            info += f", Publisher: {self.publisher}, Year: {self.year}"
        print(info)

    @classmethod
    def from_details(cls, title, author, publisher=None, year=None):
        return cls(title, author, publisher, year)


class Article(Document):
    def __init__(self, title, author, conference=None, year=None):
        super().__init__(title, author)
        self.conference = conference
        self.year = year

    def display_info(self):
        info = f"Title: {self.title}, Author: {self.author}"
        if self.conference and self.year:
            info += f", Conference: {self.conference}, Year: {self.year}"
        print(info)

    @classmethod
    def from_details(cls, title, author, conference=None, year=None):
        return cls(title, author, conference, year)


def save_to_csv(file_name, data, headers):
    with open(file_name, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        for row in data:
            writer.writerow(row)


def read_from_csv(file_name):
    try:
        with open(file_name, mode='r') as file:
            reader = csv.reader(file)
            data = [row for row in reader]
        return data
    except FileNotFoundError:
        print(f"File {file_name} not found!")
        return []


if __name__ == "__main__":
    book1 = Book.from_details("To Kill a Mockingbird", "Harper Lee", "J.B. Lippincott & Co.", 1960)
    book2 = Book.from_details("Brave New World", "Aldous Huxley")
    book1.display_info()
    book2.display_info()

    article1 = Article.from_details("AI in Healthcare", "John Doe", "MedAI Conference", 2023)
    article2 = Article.from_details("Deep Learning Applications", "Jane Smith")
    article1.display_info()
    article2.display_info()

    books_data = [
        [book1.title, book1.author, book1.publisher, book1.year],
        [book2.title, book2.author, book2.publisher, book2.year],
    ]
    articles_data = [
        [article1.title, article1.author, article1.conference, article1.year],
        [article2.title, article2.author, article2.conference, article2.year],
    ]
    save_to_csv("books.csv", books_data, ["Title", "Author", "Publisher", "Year"])
    save_to_csv("articles.csv", articles_data, ["Title", "Author", "Conference", "Year"])

    print("\nBooks from CSV:")
    books_csv = read_from_csv("books.csv")
    for row in books_csv:
        print(row)

    print("\nArticles from CSV:")
    articles_csv = read_from_csv("articles.csv")
    for row in articles_csv:
        print(row)
```
